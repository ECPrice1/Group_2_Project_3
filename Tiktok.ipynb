```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load 
Tiktok_2019 = Path("Resources/Tiktok_songs_2019.csv")
Tiktok_2020 = Path("Resources/Tiktok_songs_2020.csv")
Tiktok_2021 = Path("Resources/Tiktok_songs_2021.csv")
Tiktok_2022 = Path("Resources/Tiktok_songs_2022.csv")

# Read csv files and store into Pandas DataFrames
data_2019 = pd.read_csv(Tiktok_2019)
data_2020 = pd.read_csv(Tiktok_2020)
data_2021 = pd.read_csv(Tiktok_2021)
data_2022 = pd.read_csv(Tiktok_2022)
```

```python
data_2022["Year"] = 2022
data_2022
```

```python
data_2021["Year"] = 2021
data_2021
```

```python
data_2020["Year"] = 2020
data_2020
```

```python
data_2019["Year"] = 2019
data_2019
```

```python
data_list = [data_2019, data_2020, data_2021, data_2022] 
tik_tok_data = pd.concat(data_list).reset_index(drop = True)
tik_tok_data
```

```python
tik_tok_data = tik_tok_data.drop(columns=['time_signature', 'duration_ms', 'mode'])
tik_tok_data
```
